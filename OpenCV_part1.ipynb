### step by step



```python
import cv2
import matplotlib.pyplot as plt
```

##### reading an image

```python
img = cv2.imread(r'C:\Users\shama\Desktop\image-preprocessing-Shamah1\images\blob-09636982-a21a-494b-bbe4-3692c2720ae3.jpeg', 1)#Read the image in color
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Convert BGR to RGB for correct color display in matplotlib
plt.imshow(img) # Display the image
plt.axis('off') # Hide the axes
plt.show()
```

##### resizing an image

```python
resized = cv2.resize(img, (300 ,300)) # Resize the image to 300x300 pixels
plt.imshow(resized) # Display the resized image
plt.axis('off') # Hide the axes
plt.show()
```

##### Flip an image

```python
flip = cv2.flip(resized, 0) # Flip the image horizontally
plt.imshow(flip) # Display the flipped image
plt.axis('off') # Hide the axes
plt.show()
```

##### Rotate an image

```python
rotation = cv2.rotate(resized,2) # 0 -> Rotate the image by 90 degrees clockwise, 1 -> Rotate by 90 degrees counter-clockwise, 2 -> Rotate by 180 degrees
plt.imshow(rotation) # Display the flipped image
plt.axis('off') # Hide the axes
plt.show()
```

##### Rotation with a specific angle

```python
(h, w) = img.shape[0:2]
angle = 45
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0) # Create a rotation matrix for rotating the image by 45 degrees around its center
rotation = cv2.warpAffine(img, rotation_matrix, (w, h), flags=cv2.INTER_NEAREST) # Rotate the image by 45 degrees around its center
plt.imshow(rotation) # Display the rotated image
plt.axis('off') # Hide the axes
plt.show()
```

#### canny edge detection

```python
canny = cv2.Canny(img,25,50) # Apply Canny edge detection to the image (50 is the lower threshold, 20 is the upper threshold)
plt.imshow(canny, cmap='gray') # Display the Canny edge detected image
plt.axis('off') # Hide the axes
plt.show()
```

##### Drwaing circle

```python
circle = cv2.circle(img.copy(), center= (center), radius=50, color=(255, 0, 0), thickness=2) # Draw a circle on the image at the center with a radius of 50 pixels
plt.imshow(circle) # Display the image with the circle
plt.axis('off') # Hide the axes
plt.show()
```

```python
circle = cv2.circle(img.copy(), center= (center), radius=350, color=(255, 0, 0), thickness=2) # Draw a circle on the image at the center with a radius of 50 pixels
plt.imshow(circle) # Display the image with the circle
plt.axis('off') # Hide the axes
plt.show()
```

```python
Rectangular = cv2.rectangle(img.copy(), pt1=(400, 650), pt2=(1000, 50), color=(0, 255, 0), thickness=2) # Draw a rectangle on the image from point (50, 50) to point (250, 250)
plt.imshow(Rectangular) # Display the image with the Rectangular
plt.axis('off') # Hide the axes
plt.show()
```
